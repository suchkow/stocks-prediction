```python
import pandas as pd
```

#### Twitter Data Preprocessing

```python
from datasets import load_dataset
from datetime import datetime
dataset = load_dataset('mjw/stock_market_tweets', split='train')
```

```python
df_hf = pd.DataFrame(dataset)
```

```python
df_hf = df_hf.drop(columns=['Unnamed: 0', 'tweet_id', 'writer', 'comment_num',  'retweet_num', 'like_num'])
df_hf = df_hf.rename({'post_date': 'date', 'body': 'tweet', 'ticker_symbol': 'ticker'}, axis='columns')
df_hf['date'] = pd.to_datetime(df_hf['date'], format='%Y-%m-%d').dt.date
df_hf
```

```python
df_kag = pd.read_csv('stock_tweets.csv')
df_kag = df_kag.drop(columns=['Company Name'])
df_kag = df_kag.rename({'Date': 'date', 'Tweet': 'tweet', 'Stock Name': 'ticker'}, axis='columns')
df_kag['date'] = pd.to_datetime(df_kag['date']).dt.date
```

```python
tickets_to_drop = [x for x in df_kag['ticker'].unique().tolist() if x not in df_hf['ticker'].unique().tolist()]
print(tickets_to_drop)
```

```python
for ticket in tickets_to_drop:
    df_kag = df_kag[df_kag['ticker'].str.fullmatch(f'{ticket}') == False]

print(df_kag['ticker'].unique().tolist())
print(df_hf['ticker'].unique().tolist())
```

```python
df_kag
```

```python
df = pd.merge(df_hf, df_kag, how='outer')
df.sort_values('date')
df
```

```python
print('# of nulls in columns:')
print(df.isna().sum())
print('')

print(f'# of distimct tickets: {len(df["ticker"].unique().tolist())}')
print(f'# of years: {round(len(df["date"].unique().tolist()) / 365)}')
print(f'# of tweet dublicates: {sum(df["tweet"].duplicated().tolist())}')
```

```python
df = df.drop_duplicates(subset=['tweet'], keep=False)
```

```python
print('# of nulls in columns:')
print(df.isna().sum())
print('')

print(f'# of distimct tickets: {len(df["ticker"].unique().tolist())}')
print(f'# of years: {round(len(df["date"].unique().tolist()) / 365)}')
print(f'# of tweet dublicates: {sum(df["tweet"].duplicated().tolist())}')
print(f'# of observations: {df.shape[0]}')
```

```python
df.to_csv('tweets.csv', index=False)
```

### Stock Data Prepocessing

```python
df_aapl = pd.read_csv('AAPL.csv')
df_amzn = pd.read_csv('AMZN.csv')
df_goog = pd.read_csv('GOOG.csv')
df_msft = pd.read_csv('MSFT.csv')
df_tsla = pd.read_csv('TSLA.csv')

print('APPL:', df_aapl.shape)
print('AMZN:', df_amzn.shape)
print('GOOG:', df_goog.shape)
print('MSFT:', df_msft.shape)
print('TSLA:', df_tsla.shape)
```

```python
df_aapl['ticker'] = 'APPL'
df_amzn['ticker'] = 'AMZN'
df_goog['ticker'] = 'GOOG'
df_msft['ticker'] = 'MSFT'
df_tsla['ticker'] = 'TSLA'

print('APPL:', df_aapl.shape)
print('AMZN:', df_amzn.shape)
print('GOOG:', df_goog.shape)
print('MSFT:', df_msft.shape)
print('TSLA:', df_tsla.shape)

df_aapl.head(5)
```

```python
dfs = [df_aapl, df_amzn, df_goog, df_msft, df_tsla]
stocks_data = pd.concat(dfs, join='outer')

print(f'distinct stocks tickets: {stocks_data["ticker"].unique().tolist()}')
print(f'stocks data shape: {stocks_data.shape}')

stocks_data.head(5)
```

```python
stocks_data.to_csv('stocks.csv', index=False)
```
